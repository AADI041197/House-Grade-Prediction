```python
import pandas as pd
import numpy as np
```

```python
htrain=pd.read_csv(r'E:\pythoncode\ML Bootcamp Projects\House Grade Prediction\TRAINING.csv',header=0,index_col=0)
```

```python
htrain.head()
```

```python
htrain.shape
```

```python
htrain.isnull().sum()
```

```python
htrain.columns
```

```python
htrain.dtypes
```

```python
#replace the missing values with mean values
for value in ['Troom', 'Nbedrooms', 'Nbwashrooms', 'Twashrooms','Roof(Area)', 'Lawn(Area)','API']:
    htrain[value].fillna(htrain[value].mean(),inplace=True)
```

```python
htrain.isnull().sum()
```

```python
htrain.roof.value_counts()
```

```python
htrain.roof=htrain.roof.map({"yes":"YES","no":"NO","NO":"NO","YES":"YES"})
htrain.head()
```

```python
#replace the missing values with mode values
for value in ['roof']:
    htrain[value].fillna(htrain[value].mode()[0],inplace=True)
```

```python
htrain.roof.mode()[0]
```

```python
htrain.isnull().sum()
```

```python
htrain.dtypes
```

```python
htrain['EXPECTED'] = htrain['EXPECTED'].str.replace('$', '').astype(int)
```

```python
htrain.head()
```

```python
htrain.Grade.value_counts()
```

```python
colname=['roof','Grade']
```

```python
from sklearn import preprocessing
le=preprocessing.LabelEncoder()

for x in colname:
    htrain[x]=le.fit_transform(htrain[x])
```

```python
htrain.head()
```

```python
htrain.shape
```

```python
for x in colname:
    htrain[x]=le.fit_transform(htrain[x])
    le_name_mapping = dict(zip(le.classes_, le.transform(le.classes_)))
    print('Feature', x)
    print('mapping', le_name_mapping)
```

```python
neworder = ['Area(total)', 'Troom', 'Nbedrooms', 'Nbwashrooms', 'Twashrooms','roof',
       'Roof(Area)', 'Lawn(Area)', 'Nfloors', 'API', 'ANB','EXPECTED','Grade']
htrain=htrain.reindex(columns=neworder)
```

```python
htrain.head()
```

```python
X = htrain.values[:,:-1]
Y = htrain.values[:,-1]
```

```python
X.shape
```

```python
Y.shape
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaler.fit(X)

X = scaler.transform(X)
print(X)
```

```python
from sklearn.model_selection import train_test_split

#Split the data into test and train
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3,
random_state=10)
```

```python
#predicting using the KNeighbors_Classifier
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
model_KNN=KNeighborsClassifier(n_neighbors=19, metric='euclidean')
#fit the model on the data and predict the values
model_KNN.fit(X_train,Y_train)

Y_pred=model_KNN.predict(X_test)
print(list(zip(Y_test,Y_pred)))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score,\
classification_report

cfm=confusion_matrix(Y_test,Y_pred)
print(cfm)

print("Classification report: ")

print(classification_report(Y_test,Y_pred))

acc=accuracy_score(Y_test, Y_pred)
print("Accuracy of the model: ",acc)
```

```python
from sklearn import svm
svc_model=svm.SVC(kernel='rbf',C=1.0,gamma=0.1)
svc_model.fit(X_train, Y_train)
Y_pred=svc_model.predict(X_test)
#print(list(Y_pred))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score,\
classification_report

cfm=confusion_matrix(Y_test,Y_pred)
print(cfm)

print("Classification report: ")

print(classification_report(Y_test,Y_pred))

acc=accuracy_score(Y_test, Y_pred)
print("Accuracy of the model: ",acc)
```

```python
#predicting using the Decision_Tree_Classifier
from sklearn.tree import DecisionTreeClassifier

model_DecisionTree=DecisionTreeClassifier(random_state=10)

#fit the model on the data and predict the values
model_DecisionTree.fit(X_train,Y_train)
Y_pred=model_DecisionTree.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score,\
classification_report

cfm=confusion_matrix(Y_test,Y_pred)
print(cfm)

print("Classification report: ")

print(classification_report(Y_test,Y_pred))

acc=accuracy_score(Y_test, Y_pred)
print("Accuracy of the model: ",acc)
```

```python
#predicting using the Random_Forest_Classifier
from sklearn.ensemble import RandomForestClassifier

model_RandomForest=RandomForestClassifier(100,random_state=10)

#fit the model on the data and predict the values
model_RandomForest.fit(X_train,Y_train)

Y_pred=model_RandomForest.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score,\
classification_report

cfm=confusion_matrix(Y_test,Y_pred)
print(cfm)

print("Classification report: ")

print(classification_report(Y_test,Y_pred))

acc=accuracy_score(Y_test, Y_pred)
print("Accuracy of the model: ",acc)
```
